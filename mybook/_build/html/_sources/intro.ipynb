### The Framework

We can understand a given application in the world of AI/ML as very complex systems of input-process-output put together.


For each use case, we have 3 key aspects of this **input-process-output system**:

**(1) data types**

The type of data being fed to the system and being outputted by the system. For complex systems with intermediary processes, you can also have a third data type being used between the data type fed and the data type outputted.

**(2) architectures**

The structural design of the model that processes and transforms the data. Architectures define the flow that the data being transformed will follow — how data is processed. Think about them as the blueprint of what steps data goes through.

**(3) techniques**

The methods, algorithms, and strategies used to train, optimize, and apply the architectures. They are the specific features that together defines an architecture.

So a system is the sum of all data types, architectures and techniques involved in its constitution.

This project was made to capture these different systems used nowadays. That is, of course, a non-exhaustive list — mainly because the AI/ML field keep advancing. However, I hope it can help you identify where new articles and discoveries belong, understanding how they impact the field and for which use cases that new advancement is useful for.

At the end of the day it is all about what type of data is given as input, the process through which we transform and manipulate this data and which new data is generated from this process. 

That can get complex simply because you can add as much pieces as you want in the middle of this “process” in a input-process-output system. 

For example, in the middle of the process you could add another structure that get the original raw data slightly transformed and send it as input to another process that transforms it a bit and gives the transformed data back as output for your original process so it can keep its work and give a final output.

### Approach to content and topics structure

It is hard to come up with a good mental model or conceptual framework to separate AI/ML concepts nowadays, there is so much overlap. So the approach followed in this project was to divide them by the relevant data type. Then, if "text" is the main data type in the context (ex. for Transformers architecture or GPT-3), the topic will be under "Language Models". In the same way, if "image" is the main data type (ex. for CNNs or Stable Diffusion), the topic will be under "Image Models".

That's an attempt to make it easier to the reader to search for content related to the data type he is working with.

Note that some architectures and techniques, once discovered for one specific application, are normally used or adapted for other use cases (exploring different data types and combinations with other architectures and techniques). 

I’ll try my best to point this nuances when describing concepts, models and new discoveries throughout this text, but be aware of them. For example, sometimes I’ll add a concept under a specific topic that is most related to it, like RAG under Language Models, just because Languages Models (and the data type text) is the most common context for RAG techniques to be applied to. However, this concept can also be used within other contexts and data types. The goal is to make these nuances as clear as possible.

Also, I’ll try to go deeper in the architectures, techniques and concepts that are the most famous now, only briefly mentioning outdated techniques and architectures for focus reasons. 

The goal of this project is to cover the greatest number of topics that are relevant to AI applications nowadays, ensuring that you at least get to know the topic exists and why it is useful. That means I can’t go in depth into everything, but also that I can mention all the relevant topics and you can go in depth later into the ones you find out to be more relevant to your use case. I’ll be sure to mention the papers that come up with the architectures, techniques and concepts, as well as blogs or videos on the topics.

I also added a special chapter named "Business", in which I explore and discuss topics related to the business opportunities of applying these concepts to create products and companies.
